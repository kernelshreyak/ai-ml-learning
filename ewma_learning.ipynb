```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv("sample_datasets/delhi_temperatures.csv", parse_dates=['Date'])

# remove unnecessary columns
df = df.drop(columns=['_id'])
df_2019 = df[df['Date'].dt.year == 2019]

df_2019.head(50)
```

```python
# plot temperature with date
plt.plot(df_2019['Date'], df_2019['Rain'], label="Original Temperature")
```
