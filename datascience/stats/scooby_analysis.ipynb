```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../datasets/scoobydoo.csv")
df = df.drop(columns=['index'])
df.head()
```

```python
# df.iloc[[0,1]]
df['series_name'].unique()
```

```python
df.loc[df['series_name'] == "Laff-a-Lympics"].shape
```

```python
# Convert the imdb rating to numeric (handling errors for any non-numeric values)
df['imdb'] = pd.to_numeric(df['imdb'], errors='coerce')

# Group by series_name and calculate the average IMDB rating for each series
average_imdb_by_series = df.groupby('series_name')['imdb'].mean().dropna()

# Plotting the bar graph
plt.figure(figsize=(10, 6))
average_imdb_by_series.plot(kind='barh')
plt.title('Average IMDB Rating of Each Scooby-Doo Series')
plt.xlabel('Series Name')
plt.ylabel('Average IMDB Rating')
plt.xticks(ha='right')
plt.tight_layout()
plt.show()
```
