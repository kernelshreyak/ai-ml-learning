```python
lr_loss = np.vectorize(lambda y, y_hat: -(y*np.log(y_hat) + (1-y)*np.log(1-y_hat)))
# some sample predictions
y_hat = [0.1,0.5,0.6,0.75,0.9,0.992,0.9998]
y = np.ones(len(y_hat))
plt.plot(range(len(y_hat)), lr_loss(y,y_hat))
```
